# Netflix Genre Popularity Analysis (2019 vs 2021)

This project aims to analyze the popularity of Netflix content by genre in the years 2019 and 2021. The dataset used includes movies and TV shows available on Netflix, and we will be focusing on visualizing the distribution of genres across these two years.

## Objectives:
- Filter Netflix data to focus on content released in 2019 and 2021.
- Group content by genre and year to identify trends.
- Visualize the genre distribution with a stacked bar chart to compare popularity across years.

  

## Importing Libraries and Loading the Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('netflix_series_10_columns_data.csv')

df.head()
```

In this section, we load the Netflix dataset, which contains various columns like `Title`, `Genre`, `Release Year`, and more. The dataset will be filtered and analyzed to focus on genre-based trends for 2019 and 2021.



## Filtering Data for 2019 and 2021

```python
df_filtered = df[df['Release Year'].isin([2019, 2021])]

df_filtered.head()
```

We filter the dataset to focus on content released in the years 2019 and 2021. This will help us compare genre popularity between these two years.



## Grouping by Genre and Year

```python
genre_year = df_filtered.groupby(['Release Year', 'Genre']).size().unstack().fillna(0)

genre_year.head()
```

After filtering the data, we group it by `Release Year` and `Genre` to count the number of titles released in each genre for both 2019 and 2021.



## Visualizing Genre Popularity

```python
genre_year.plot(kind='bar', figsize=(10, 7), stacked=True, colormap='coolwarm')

plt.title('Genre Popularity on Netflix (2019 vs 2021)', fontsize=16)
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)

plt.legend(title="Genres", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()

plt.show()
```

The chart above compares the popularity of Netflix genres in 2019 and 2021. Each genre is represented by a different color in the stacked bar chart, allowing for an easy comparison of how many titles in each genre were released in these two years.



## Insightful Observations

### Observations:
- From the visualization, its clear that certain genres like Fantasy and Documentary saw a significant increase in 2021 compared to 2019.
- Other genres, such as Mystery and Drama, remained relatively stable across both years.
- This analysis could suggest changing viewer preferences, potentially influenced by global events like the pandemic in 2020.

Further analysis can explore whether certain genres perform better in terms of audience reception (e.g., ratings) and how these trends might continue in subsequent years.
